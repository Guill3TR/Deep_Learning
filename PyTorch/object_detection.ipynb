## **OBJECT DETECTION PROJECT USING DETECTO PACKAGE**

The Project consists of a dataset for object detection, formed by 240 images in train folder and 60 images in test folder.

The contents are 2 differents types of fruit:

- Apple
- Banana

### Packages

```python
!pip install detecto
from detecto import core, utils
from detecto.visualize import show_labeled_image
from torchvision import transforms
import numpy as np
```

### Data Preparation

```python
path_images = '/content/images'
path_train_labels = '/content/train_labels'
path_test_labels = '/content/test_labels'
```

```python
custom_transforms = transforms.Compose([
    transforms.ToPILImage(),
    transforms.Resize((50)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(15),
    transforms.ToTensor(),
    utils.normalize_transform()
])
```

```python
trained_labels = ['apple', 'banana']

train_dataset = core.Dataset(image_folder=path_images, label_data=path_train_labels, transform=custom_transforms)
test_dataset = core.Dataset(image_folder=path_images, label_data=path_test_labels, transform=custom_transforms)

train_loader = core.DataLoader(train_dataset, batch_size=2, shuffle=False)
test_loader = core.DataLoader(test_dataset, batch_size=2, shuffle=False)
```

### Model

```python
model = core.Model(trained_labels) #core from detecto
```

```python
losses = model.fit(train_loader, test_dataset, epochs=10, verbose=True)
```

```python
test_image_path = 'images/apple_77.jpg'
test_image = utils.read_image(test_image_path)
pred = model.predict(test_image)
labels, boxes, scores = pred #unpacking the prediction results
show_labeled_image(test_image, boxes, labels)
```

Load a test image, make a prediction using a previously trained model, and then visualise the test image with the predictions overlaid

```python
conf_threshold = 0.7 #minimum confidence required for a prediction to be considered valid
filtered_indices = np.where(scores > conf_threshold)
filteres_scores = scores[filtered_indices]
filtered_boxes = boxes[filtered_indices]
num_list = filtered_indices[0].tolist() #we only need the indices in one dimension
filtered_labels = [labels[i] for i in num_list]
show_labeled_image(test_image, filtered_boxes, filtered_labels)
```

### Get Predictions

```python
y_test_pred = []
import torch
with torch.no_grad():
    for j, data in enumerate(test_loader):
        image, label = data #unpacks the test batch data in image and label variables
        output = model.predict(image)
        y_test_pred.extend(output) #the model's predictions are added to the list
        break
```
